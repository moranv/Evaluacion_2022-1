```python
import pandas as pd
import requests
import json
```

```python
r=pd.read_json('https://raw.githubusercontent.com/ComputationalMethods/Evaluacion_2022-1/main/data/lista_pares.json')
r
```

```python
#B. 1) Construcción de la matriz
```

```python
X = r.lista[65]
Y = r.pares[65]['ψ']
lista0 =[]
for  i in X:
    lista = []
    for j in X:
        if [i,j] in r.pares[65]['ψ']:
            lista.append(round(random.random(),2))
        else:
            lista.append(0)
    
    lista0.append(lista)
Y = np.array(lista0)
Z = sp.Matrix(Y)
Z
```

```python
# 2) Cálculo del determinante
```

```python
det = np.linalg.det(Y)
print(r'El determinante de la matriz M, es: %.14f'%det)
```

```python
# 3) Autovalores y autovectores
```

```python
autovalor, autovector = np.linalg.eig(Y)
A_diag=np.diag(autovalor)

print('Autovalores: \n',np.sort( np.abs(autovalor)))
v_abs=abs(autovector)
np.c_[ tuple( [ np.c_[v_abs[:,i]]    for i in range(8) ] ) ]
U=np.c_[ tuple( [ np.c_[v_abs[:,i]]    for i in np.abs(autovalor).argsort() ] ) ]
print('Autovectores ordenados: \n')
sp.Matrix((U.transpose()@Y@U).round(14))
```

```python
sp.Matrix(np.diag(( U.transpose()@Y@U).round(14)))==sp.Matrix((np.sort(np.abs(autovalor).round(14))))
```

```python
# 4) Autovalores degenerados en valor absoluto
```

```python
sp.Matrix(np.diag(( U.transpose()@Y@U).round(14)))==sp.Matrix((np.sort(np.abs(autovalor).round(14))))

# Los autovalores no son degenerados alrededor de 14 cifras significativas 
```
